**Exercice 2   — IRIS Data Classification using CART Decision Tree**

1. Charger le dataset “Iris.csv”.

```python
import pandas as pd

iris_data = pd.read_csv('IRIS.csv')
```

2. Afficher les données

```python
iris_data.head()
```

-longueur du sépale : La longueur du sépale en centimètres.

-largeur du sépale : La largeur du sépale en centimètres.

-longueur du pétale : La longueur du pétale en centimètres.

-largeur du pétale : La largeur du pétale en centimètres.

-espèce : L'espèce d'iris. Cette colonne a trois valeurs possibles : "Iris-setosa", "Iris-versicolor" et "Iris-virginica".

3. Proposer deux modèles CART.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
# Diviser les données en features et target
X = iris_data.drop("species", axis=1)
y = iris_data["species"]

```

Modèle 1 : CART avec critère de Gini et profondeur max = 50

```python

model1 = DecisionTreeClassifier(criterion="gini", max_depth=50)
model1.fit(X, y)

```

Modèle 2 : CART avec critère d'entropie et profondeur max = 4

```python

model2 = DecisionTreeClassifier(criterion="entropy", max_depth=4)
model2.fit(X, y)
```

4. Visualiser les résultats graphiquement. 

```python
import matplotlib.pyplot as plt
# Couleurs pour chaque espèce
colors = {"Iris-setosa": "blue", "Iris-versicolor": "green", "Iris-virginica": "red"}

# Visualiser les arbres de décision avec des couleurs différentes
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5), dpi=300)
plot_tree(model1, feature_names=X.columns, class_names=y.unique(), ax=ax[0], filled=True, rounded=True, proportion=True, impurity=False, fontsize=6, node_ids=True, precision=2)
ax[0].set_title("CART avec critère de Gini et profondeur max = 50", fontsize=8)

plot_tree(model2, feature_names=X.columns, class_names=y.unique(), ax=ax[1], filled=True, rounded=True, proportion=True, impurity=False, fontsize=6, node_ids=True, precision=2)
ax[1].set_title("CART avec critère d'entropie et profondeur max = 4", fontsize=8)

plt.show()

```
