**Exercice 3   — Credit Card Fraud detection using Decision Tree**

1.  Charger les données

```python
import pandas as pd

data = pd.read_csv("creditcard.csv")
```

2. Afficher quelques données:

```python
data.head()
```

3.  Proposer deux modèles basés sur les arbres de décisions.

Modèle 1: Utiliser l'arbre de décision par défaut

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# Diviser les données en ensembles de formation et de test
X_train, X_test, y_train, y_test = train_test_split(data.drop('Class', axis=1), data['Class'], test_size=0.3, random_state=42)

# Créer un arbre de décision
dt = DecisionTreeClassifier()

# Entraîner l'arbre de décision sur l'ensemble de formation
dt.fit(X_train, y_train)

# Évaluer l'arbre de décision sur l'ensemble de test
score = dt.score(X_test, y_test)
print("Score de l'arbre de décision par défaut: {:.2f}%".format(score*100))
```

Modèle 2: Utiliser un arbre de décision avec une profondeur limitée à 3

```python
# Créer un arbre de décision avec une profondeur limitée à 3
dt_depth3 = DecisionTreeClassifier(max_depth=3)

# Entraîner l'arbre de décision sur l'ensemble de formation
dt_depth3.fit(X_train, y_train)

# Évaluer l'arbre de décision sur l'ensemble de test
score_depth3 = dt_depth3.score(X_test, y_test)
print("Score de l'arbre de décision avec une profondeur limitée à 3: {:.2f}%".format(score_depth3*100))
```

4.  Visualiser les résultats comparatifs graphiquement.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Afficher l'arbre de décision par défaut
plt.figure(figsize=(10,8))
plot_tree(dt, filled=True)
plt.title("Arbre de décision par défaut")
plt.show()

# Afficher l'arbre de décision avec une profondeur limitée à 3
plt.figure(figsize=(10,8))
plot_tree(dt_depth3, filled=True)
plt.title("Arbre de décision avec une profondeur limitée à 3")
plt.show()
```
