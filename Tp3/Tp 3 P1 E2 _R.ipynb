3. Créer puis visualiser l’arbre. 

```r
library(rpart)
model <- rpart(Species ~ ., data = iris, method = "class")
plot(model)
text(model)
```

4. Visualiser la qualité de la classification.

```r
library(rpart.plot)
rpart.plot(model, type = 4, extra = 101)
```
