Exercice 2

LungCapData <- read.table(file.choose(), header = T, sep="\t") : Cette ligne de code permet de charger les données du fichier dans un objet appelé LungCapData. La fonction file.choose() permet de sélectionner le fichier à partir de l'interface graphique. header=T indique que le fichier contient une ligne d'en-tête et sep="\t" indique que les colonnes sont séparées par des tabulations.

attach(LungCapData) : Cette fonction permet de joindre les données de LungCapData à l'environnement global.

names(LungCapData) : Cette ligne permet d'afficher les noms des colonnes dans le fichier.

class(Age) : Cette ligne permet d'afficher la classe de la variable Age.

class(Smoke) : Cette ligne permet d'afficher la classe de la variable Smoke.

levels(Smoke) : Cette ligne permet d'afficher les niveaux de la variable Smoke.

model1 <- lm(LungCap ~ Age + Height) : Cette ligne permet de créer un modèle linéaire (lm) où la capacité respiratoire (LungCap) est expliquée par l'âge (Age) et la taille (Height).

summary(model1) : Cette ligne permet d'afficher un résumé du modèle model1. Les résultats incluent les coefficients de régression, les erreurs standards, les statistiques de test et les valeurs p.

cor(Age, Height, method="pearson") : Cette ligne permet de calculer le coefficient de corrélation de Pearson entre Age et Height.

confint(model1, coef.level=0.95) : Cette ligne permet de calculer les intervalles de confiance à 95% pour les coefficients du modèle model1.

model2 <- lm(LungCap ~ Age + Height + Smoke + Gender + Caesarean) : Cette ligne permet de créer un modèle linéaire (lm) où la capacité respiratoire (LungCap) est expliquée par l'âge (Age), la taille (Height), le tabagisme (Smoke), le genre (Gender) et le mode d'accouchement (Caesarean).

summary(model2) : Cette ligne permet d'afficher un résumé du modèle model2. Les résultats incluent les coefficients de régression, les erreurs standards, les statistiques de test et les valeurs p.

plot(model2) : Cette ligne permet de tracer un graphe des résidus du modèle model2 pour détecter des schémas dans les résidus.
