charger les données du fichier dans un objet appelé LungCapData. La fonction file.choose() permet de sélectionner le fichier à partir de l'interface graphique. header=T indique que le fichier contient une ligne d'en-tête et sep="\t" indique que les colonnes sont séparées par des tabulations

```r
LungCapData <- read.table("LungCapData.txt", header = T, sep="\t")
```

```r
attach(LungCapData)
```

créer un modèle linéaire (lm) où  LungCap est expliquée par Age et Height

```r
model1 <- lm(LungCap ~ Age + Height)
```

affiche un résumé des résultats du modèle 1

```r
summary(model1) 
```

calculer le coefficient de corrélation de Pearson entre Age et Height

```r
cor(Age, Height, method="pearson")
```

calculer les intervalles de confiance à 95% pour les coefficients du modèle model1

```r
confint(model1, coef.level=0.95)
```

créer un modèle linéaire (lm) où LungCap est expliquée par Age, Height, Smoke, Gender et Caesarean

```r
model2 <- lm(LungCap ~ Age + Height + Smoke + Gender + Caesarean)
```

affiche un résumé des résultats du modèle 2

```r
summary(model2)
```

tracer un graphe des résidus du modèle model2
