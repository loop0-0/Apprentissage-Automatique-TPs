Exercice 2 :

1. Téléchargez les données (MNIST).

```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

2. Divisez les données en ensembles d’apprentissage et de test (MNIST). 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.25, random_state=42)
```

3. Affichez les images et les étiquettes (MNIST).

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))
for i, ax in enumerate(axs.ravel()):
    ax.imshow(x_train[i], cmap='gray')
    ax.axis('off')
    ax.set_title(str(y_train[i]))
plt.tight_layout()
plt.show()
```

4. Créez un Modèle d’apprentissage (MNIST) :

Étape 1. Importez le modèle que vous souhaitez utiliser.

```python
from sklearn.linear_model import LogisticRegression
```

Étape 2. Créez une instance du modèle.

```python
model = LogisticRegression(max_iter=1000)
```

Étape 3. Formation du modèle sur les données, stockage des informations apprises à partir des 
données.

```python
model.fit(x_train.reshape(-1, 784), y_train)
```

Étape 4. Prédisez les étiquettes pour les nouvelles données (nouvelles images).

```python
y_pred = model.predict(x_test.reshape(-1, 784))
```

9. Mesurez la performance du modèle (MNIST).

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

10. Affichez les images mal classées avec des étiquettes prédites (MNIST)

```python
import numpy as np

misclassified_idx = np.where(y_test != y_pred)[0]

fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))
for i, ax in enumerate(axs.ravel()):
    idx = misclassified_idx[i]
    ax.imshow(x_test[idx], cmap='gray')
    ax.axis('off')
    ax.set_title(f"True: {y_test[idx]}\nPred: {y_pred[idx]}")
plt.tight_layout()
plt.show()
```
